# Convert columns to numbers + create t0

```python
def prepare_time_columns(df):
    df = df.copy()

    df["t"] = pd.to_numeric(df["t"], errors="coerce")
    df["measureStartTime"] = pd.to_numeric(df["measureStartTime"], errors="coerce")
    df["full"] = pd.to_numeric(df["full"], errors="coerce")

    df = df.dropna(subset=["t", "measureStartTime", "full"])

    df["t0"] = df["t"] - df["measureStartTime"]
    return df

df_g_r1 = prepare_time_columns(df_g_r1)
df_g_r2 = prepare_time_columns(df_g_r2)

df_p_r1 = prepare_time_columns(df_p_r1)
df_p_r2 = prepare_time_columns(df_p_r2)

df_r_r1 = prepare_time_columns(df_r_r1)
df_r_r2 = prepare_time_columns(df_r_r2)

df_0_r1 = prepare_time_columns(df_0_r1)
df_0_r2 = prepare_time_columns(df_0_r2)

df_r_suitcase_r1 = prepare_time_columns(df_r_suitcase_r1)
df_r_suitcase_r2 = prepare_time_columns(df_r_suitcase_r2)
df_r_suitcase_r3 = prepare_time_columns(df_r_suitcase_r3)
df_r_suitcase_r4 = prepare_time_columns(df_r_suitcase_r4)

df_clamp_pp_r1 = prepare_time_columns(df_clamp_pp_r1)
df_clamp_G_r1 = prepare_time_columns(df_clamp_G_r1)
df_clamp_R_r1 = prepare_time_columns(df_clamp_R_r1)
df_clamp_ultra_r1 = prepare_time_columns(df_clamp_ultra_r1)
df_clamp_white_r1 = prepare_time_columns(df_clamp_white_r1)
df_clamp_white_noLSR_r1 = prepare_time_columns(df_clamp_white_noLSR_r1)
df_clamp_macha_r1 = prepare_time_columns(df_clamp_macha_r1)
df_clamp_macha_noLSR_r1 = prepare_time_columns(df_clamp_macha_noLSR_r1)


```

# Get table with key values

```python
t0_targets = [
    60_000,        # 1 min
    300_000,       # 10 min
    3_600_000,     # 1 hour
    7_200_000,     # 2 hours
    10_800_000,    # 3 hours
    14_400_000,    # 4 hours
    18_000_000,    # 5 hours
    21_600_000,    # 6 hours
    25_200_000,    # 7 hours
    28_800_000     # 8 hours
]
```

```python
def full_local_avg(df, target_t0, n=10):
    df = df.sort_values("t0")

    before = df[df["t0"] <= target_t0].tail(n)
    after  = df[df["t0"] >  target_t0].head(n)

    combined = pd.concat([before, after])

    return combined["full"].mean()
```

```python
dfs = {
    "Green Run 1": df_g_r1,
    "Green Run 2": df_g_r2,
    "Phillips GID Run 1": df_p_r1,
    "Phillips GID Run 2": df_p_r2,
    "Red Run 1": df_r_r1,
    "Red Run 2": df_r_r2,
    #"PP Run 1": df_0_r1,
}

table = pd.DataFrame(
    {name: [full_local_avg(df, t0) for t0 in t0_targets]
     for name, df in dfs.items()},
    index=t0_targets
)

table.index.name = "t0 (ms)"
table
```

```python
table.insert(
    0,
    "Time",
    ["1 min","10 min","1 hr","2 hr","3 hr","4 hr","5 hr","6 hr","7 hr","8 hr"]
)

table
```

```python
# x-axis labels (from the table)
x_labels = table["Time"]
x_positions = range(len(x_labels))

plt.figure(figsize=(10, 6))

for col in table.columns[1:]:  # skip "Time"
    plt.plot(
        x_positions,
        table[col],
        marker="o",
        label=col
    )

plt.xticks(x_positions, x_labels)
plt.xlabel("Time")
plt.ylabel("Full")
plt.title("Full value vs Time")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Smoothen data by calculating rolling average

def smooth_df(df, window=5):
    df = df.copy()
    
    # select numeric columns only
    numeric_cols = df.select_dtypes(include="number").columns
    
    # apply rolling mean
    df[numeric_cols] = (
        df[numeric_cols]
        .rolling(window=window, min_periods=1)
        .mean()
    )
    
    return df

df_g_r1 = smooth_df(df_g_r1, window=10)
df_g_r2 = smooth_df(df_g_r2, window=10)

df_p_r1 = smooth_df(df_p_r1, window=10)
df_p_r2 = smooth_df(df_p_r2, window=10)

df_r_r1 = smooth_df(df_r_r1, window=10)
df_r_r2 = smooth_df(df_r_r2, window=10)


```
